11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?
#### Solution:
#### In for loops there is no difference between range(10), range(0, 10), and range(0, 10, 1) the produce the same result however when range(10) is written it takes 0 as lower bound and step size as 1 by default and when range(0,10) is written it takes step size as 1 by default.


```python
# 12. Write a short program that prints the numbers 1 to 10 using a for loop. 
#     Then write an equivalent program that prints the numbers 1 to 10 using a while loop.

# Solution:
# i) prints the numbers 1 to 10 using a for loop:

for i in range(1,11):
    print(i)

# ii) Then write an equivalent program that prints the numbers 1 to 10 using a while loop
num = 1
while num <= 10:
    print(num)
    num = num + 1
```
