17. How do you distinguish between copy.copy() and copy.deepcopy()?
#### Solution:
* copy and deep copy is used to create a copy of objects.

#### Shallow copy:
* The copy() function creates a shallow copy of an object.
* shallow copy of an object is a new object that is a copy of the original object's content, but it doesn't recursively copy nested objects. Instead, it creates references to the same nested objects as the original. 
* Shallow copying is useful when you want to duplicate the top-level structure of a complex object, but you're okay with shared references to nested objects.

#### Deep copy:
* Thedeepcopy() function creates a deep copy of an object.
* A deep copy creates an entirely new object hierarchy, recursively copying all nested objects as well.
* Deep copying is useful when you want to completely isolate the copied object from the original, including all nested objects.
